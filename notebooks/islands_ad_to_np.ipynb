```python
import os
import glob
import sys
import zlib
import pathlib
import tempfile
import subprocess
import json
import numpy as np

from binascii import unhexlify
from xml.etree import ElementTree
from struct import unpack

sys.path.append("../src")

from anno_designer import read_ad_json
```

```python
island_glob = os.path.join("..", "resources", "island_outlines", "*.ad")
output_dir = os.path.join("..", "resources", "islands")
```

```python
for input_path in glob.glob(island_glob):
    island_name = pathlib.Path(input_path).stem
    output_path = os.path.join(output_dir, island_name + ".npy")
    
    print(f"Converting {input_path} \u2192 {output_path}", end="")
    with open(input_path) as f:
        input_json = json.load(f)
    island_data = read_ad_json(input_json)
    np.save(output_path, island_data)
    print(" \u2713")
```
